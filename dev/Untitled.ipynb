```python
import bw2data
from bw2calc import JacobiGMRESLCA, LCA
bw2data.projects.set_current("ecoinvent-3.10-cutoff")
db = bw2data.Database("ecoinvent-3.10.1-cutoff - regionalized")
```

```python
act = db.random()
act
```

```python
method = ('EF v3.0', 'climate change', 'global warming potential (GWP100)')
```

```python
%%time

lca = LCA(
    demand={act: 1},
    method=method
)
lca.lci()
lca.lcia()
print(lca.score)
```

```python
%%time

lca2 = JacobiGMRESLCA(
    demand={act: 1},
    method=method,
    rtol=1e-6
)
lca2.lci()
lca2.lcia()
print(lca2.score)
```
